```python
df = pd.read_csv('/content/mtcars.csv')
```

```python
df.head()
```

```python
#let's use limited columns which makes more sense for serving our purpose
# The original cols_to_use were for a different dataset. We need to use columns from mtcars.
cols_to_use = ['mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
df = df[cols_to_use]
```

```python
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
df.head()
```

```python
X = df.drop('mpg', axis=1)
y = df['mpg']
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.3, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(train_X, train_y)
```

```python
reg.score(test_X, test_y)
```

```python
reg.score(train_X, train_y)
```

```python
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha=50, max_iter=100, tol=0.1)
lasso_reg.fit(train_X, train_y)
```

```python
lasso_reg.score(test_X, test_y)
```

```python
lasso_reg.score(train_X, train_y)
```

```python
from sklearn.linear_model import Ridge
ridge_reg= Ridge(alpha=50, max_iter=100, tol=0.1)
ridge_reg.fit(train_X, train_y)
```

```python
ridge_reg.score(test_X, test_y)
```

```python
ridge_reg.score(train_X, train_y)
```

```python
from sklearn.linear_model import ElasticNet
elastic_reg = ElasticNet(alpha=50, l1_ratio=0.5, max_iter=100, tol=0.1)
elastic_reg.fit(train_X, train_y)
```

```python
elastic_reg.score(test_X, test_y)
```

```python
elastic_reg.score(train_X, train_y)
```
